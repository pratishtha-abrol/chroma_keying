```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from os.path import isfile, join
```

## Videos to images

```python
video = cv2.VideoCapture('video.mp4')
os.mkdir('./images')
```

Find video used [here](https://youtu.be/QC8iQqtG0hg)

```python
def decomp(sec):
    video.set(cv2.CAP_PROP_POS_MSEC, sec*1000)
    has_frames, image = video.read()
    if has_frames:
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.show()
        cv2.imwrite('./images/image'+str(count)+'.jpg', image)
    return has_frames
```

```python
sec = 0
frame_rate = 0.5
count=1
success = decomp(sec)
while success:
    count += 1
    sec += frame_rate
    sec = round(sec, 2)
    success = decomp(sec)
```

## Images to Video

```python
imgpath = './images/'
videopath = './output_video.avi'
fps = 2
```

```python
frame_arr = []
files = [f for f in os.listdir(imgpath) if isfile(join(imgpath, f))]
files.sort(key = lambda x: x[5:-4])
```

```python
for i in range(len(files)):
    filename = imgpath + files[i]
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width, height)
    frame_arr.append(img)
```

```python
out = cv2.VideoWriter(videopath, cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
```

```python
for i in range(len(frame_arr)):
    out.write(frame_arr[i])
out.release()
```
